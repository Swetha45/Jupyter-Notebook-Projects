two primary data structures of pandas
1. Series : a one-dimensional labeled array holding data of any type. such as integers, strings, Python objects etc. this can only have one data type for the whole array. 

2. Datafram : a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.
    DataFrame can have different data types for each column. e
