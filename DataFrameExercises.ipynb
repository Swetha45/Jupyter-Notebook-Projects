Modifying DataFrames

DataFrame syntax : 
pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)[source]

Series Syntax :
pandas.Series(data=None, index=None, dtype=None, name=None, copy=None, fastpath=_NoDefault.no_default)[source]

to mention column names, in SEries we use "name", but in DF we use "columns"


```python
langs = pd.Series(["English", "Hindi", "French"], name = "Language" )
Languages = pd.DataFrame(data=langs)
Languages
```

add this column to existing dataframe

```python
df1
```

```python
df1['Language'] = langs
df1
```
